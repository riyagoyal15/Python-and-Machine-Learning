```python
obj_df["gender"] = obj_df["gender"].map( 
                   {'Female':1 ,'Male':0})
obj_df["community"] = obj_df["community"].map( 
                   {'OBC':2 ,'SC':1, 'ST':0, 'FC':3})
obj_df["educ_qua"] = obj_df["educ_qua"].map( 
                   {'Graduate & Above':3 ,'Secondary':2, 'Primary':0, 'Higher Secondary':1, })
obj_df["debt_to_bank"] = obj_df["debt_to_bank"].map( 
                   {"Yes":1 ,"No":0})
obj_df["debt_to_contractors"] = obj_df["debt_to_contractors"].map( 
                   {"Yes":1 ,"No":0})
obj_df["debt_to_money_lender"] = obj_df["debt_to_money_lender"].map( 
                   {"Yes":1 ,"No":0})
obj_df["migrated_to_current_place_6_month"] = obj_df["migrated_to_current_place_6_month"].map( 
                   {"Yes":1 ,"No":0})
obj_df["fixed_asset_land"] = obj_df["fixed_asset_land"].map( 
                   {'Yes':1 ,'No':0})
obj_df["fixed_asset_cattle"] = obj_df["fixed_asset_cattle"].map( 
                   {'Yes':1 ,'No':0})
obj_df["fixed_asset_house"] = obj_df["fixed_asset_house"].map( 
                   {'Yes':1 ,'No':0})
obj_df["type_migration"] = obj_df["type_migration"].map( 
                   { 'With Family Members':1, 'As Individual': 2, "Missing Record":0})
obj_df["emp_status_before_migration"] = obj_df["emp_status_before_migration"].map( 
                   { 'Unemployed':1, 'Employed': 2, "Missing Record":0})
obj_df["work_effect_health"] = obj_df["work_effect_health"].map( 
                   {"Yes":1 ,"No":0})
obj_df["medical_faclity_workspace"] = obj_df["medical_faclity_workspace"].map( 
                   {"Yes":1 ,"No":0})
obj_df["rest_room_woman_workplace"] = obj_df["rest_room_woman_workplace"].map( 
                   {"Yes":1 ,"No":0, "Missing Record": 2})
obj_df["problems_in_behaviour_superiors"] = obj_df["problems_in_behaviour_superiors"].map( 
                   {"Yes":1 ,"No":0})
obj_df["problems_in_policies_management"] = obj_df["problems_in_policies_management"].map( 
                   {"Yes":1 ,"No":0})
obj_df["problems_in_nonredressal_grievances"] = obj_df["problems_in_nonredressal_grievances"].map( 
                   {"Yes":1 ,"No":0})
obj_df["problems_in_others"] = obj_df["problems_in_others"].map( 
                   {"Yes":1 ,"No":0})
obj_df["schemes_by_govt_emp_generation"] = obj_df["schemes_by_govt_emp_generation"].map( 
                   {"Yes":1 ,"No":0})

obj_df["schemes_by_govt_household"] = obj_df["schemes_by_govt_household"].map( 
                   {"Yes":1 ,"No":0})

obj_df["schemes_by_govt_agri_ins"] = obj_df["schemes_by_govt_agri_ins"].map( 
                   {"Yes":1 ,"No":0})

obj_df["schemes_by_govt_health_ins"] = obj_df["schemes_by_govt_health_ins"].map( 
                   {"Yes":1 ,"No":0})

obj_df["schemes_by_govt_other"] = obj_df["schemes_by_govt_other"].map( 
                   {"Yes":1 ,"No":0})

obj_df["below_poverty_line"] = obj_df["below_poverty_line"].map( 
                   {"Yes":1 ,"No":0, "Missing Record" : 2})
obj_df["health_status"] = obj_df["health_status"].map( 
                   {'Good':4 ,'Average':3, 'Very Poor':1, 'Poor': 2, 'Excellent': 5})

obj_df["quality_treatment"] = obj_df["quality_treatment"].map( 
                   {'Good':4 ,'Average':3, 'Very Poor':1, 'Poor': 2, "Missing Record":0, 'Excellent': 5})
obj_df["treat_location"] = obj_df["treat_location"].map( 
                   {'Private':3, 'Government':1, 'Self': 2, "Missing Record":0})
```
