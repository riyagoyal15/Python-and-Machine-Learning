```python
import pandas as pd
import numpy as np

```

```python
df=pd.read_csv('D:/Downloads/data.csv')
```

```python
df1=pd.read_csv('D:/Downloads/labels.csv')
```

```python
df.rename(columns = {'Unnamed: 0':'Sample_no'},inplace=True)
```

```python
df1.head()
```

```python
df1.rename(columns = {'Unnamed: 0':'Sample_no'},inplace=True)
```

```python
df3=df.merge(df1[['Sample_no', 'Class']], on = 'Sample_no', how = 'left')
```

```python
del df3['Sample_no']
```

```python
df3.head()
```

```python
y = df3.Class
X = df3.drop(['Class'],axis=1)
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=143)
```

```python
from sklearn import preprocessing
import numpy as np

# normalize the data attributes
X_train_sc = preprocessing.normalize(X_train)
X_test_sc = preprocessing.normalize(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn import tree
import graphviz
from sklearn.model_selection import cross_val_score
```

```python
print("Before OverSampling, counts of label 'BRCA ': {}".format(sum(y_train == 'BRCA')))
print("Before OverSampling, counts of label 'KIRC ': {} \n".format(sum(y_train == 'KIRC')))
print("Before OverSampling, counts of label 'LUAD ': {} \n".format(sum(y_train == 'LUAD')))
print("Before OverSampling, counts of label 'COAD ': {} \n".format(sum(y_train == 'COAD')))
print("Before OverSampling, counts of label 'PRAD ': {} \n".format(sum(y_train == 'PRAD')))

# import SMOTE module from imblearn library
from imblearn.over_sampling import SMOTE
from sklearn.metrics import DistanceMetric
sm = SMOTE(random_state = 2)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print('After OverSampling, the shape of X_train: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of y_train: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label 'BRCA ': {}".format(sum(y_train_res == 'BRCA')))
print("After OverSampling, counts of label 'KIRC ': {} \n".format(sum(y_train_res == 'KIRC')))
print("After OverSampling, counts of label 'LUAD ': {} \n".format(sum(y_train_res == 'LUAD')))
print("After OverSampling, counts of label 'COAD ': {} \n".format(sum(y_train_res == 'COAD')))
print("After OverSampling, counts of label 'PRAD ': {} \n".format(sum(y_train_res == 'PRAD')))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
gem=dt.fit(X_train_res, y_train_res)
```

```python
dt.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
tree.plot_tree(gem);
```

```python
fn=X_train.columns
```

```python

cn=['COAD', 'KIRC', 'LUAD','BRCA','PRAD']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (8,8), dpi=300)
tree.plot_tree(gem,
               feature_names=fn,
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```

```python
y_pred2 = dt.predict(X_test)
print('Accuracy of Decision Tree Classifier classifier on test set: {:.3f}'.format(dt.score(X_test, y_test)))
```

```python
confusion_matrix2 = confusion_matrix(y_test, y_pred2)
print(confusion_matrix2)
```

```python
print(classification_report(y_test,y_pred2))
```

```python
trainy_pred2 = dt.predict(X_train_res)
print('Accuracy of Decision Tree classifier on test set: {:.2f}'.format(dt.score(X_train_res, y_train_res)))

cm2 = confusion_matrix(y_train_res, trainy_pred2)
print("\n",cm2)

print("\n",classification_report(y_train_res,trainy_pred2))
```
